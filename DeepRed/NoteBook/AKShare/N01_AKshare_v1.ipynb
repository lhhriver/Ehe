```python
from threading import Thread, Lock, Semaphore

sem = Semaphore(16)  # 此变量用于控制最大并发数

stock_zh_a_hist_df_60 = pd.DataFrame()


def work(eachlist, lock):
    global stock_zh_a_hist_df_60
    sem.acquire()
    stock_zh_a_hist_tmp = pd.DataFrame()
    for code in eachlist:
        stock_zh_a_hist_ = ak.stock_zh_a_hist(symbol=code,
                                              period="daily",
                                              start_date=b_60_day,
                                              end_date=last_day,
                                              adjust="")
        stock_zh_a_hist_['代码'] = code
        stock_zh_a_hist_tmp = stock_zh_a_hist_tmp.append(stock_zh_a_hist_)

    lock.acquire()
    stock_zh_a_hist_df_60 = stock_zh_a_hist_df_60.append(stock_zh_a_hist_tmp)
    lock.release()
    sem.release()


stock_list = stock_df_ss['代码'].values.tolist()
stock_epoch = [stock_list[i:i + int(len(stock_list) / 100)] for i in range(0, len(stock_list), int(len(stock_list) / 100))]

lock = Lock()

l = []
for eachlist in stock_epoch:
    p = Thread(target=work, args=(eachlist, lock))
    l.append(p)

for p in l:
    p.start()

for p in l:
    p.join()
```

# 策略

## 上个交易日数据

```python
# 昨天数据
stock_zh_a_hist_df_last_day = stock_zh_a_hist_df_60[stock_zh_a_hist_df_60['日期'] == last_day_]
stock_zh_a_hist_df_last_day = stock_zh_a_hist_df_last_day[['代码', '开盘', '收盘']]
stock_zh_a_hist_df_last_day.columns = ['代码', '昨开', '昨收']
stock_zh_a_hist_df_last_day
```

## 60日均价

```python
# 60日均价
data_avg_60 = stock_zh_a_hist_df_60[['代码', '收盘']].groupby(['代码'], as_index=False)['收盘'].mean().round(2)
data_avg_60.columns = ['代码', '收盘_60']
data_avg_60
```

## 筛选股票

```python
print(stock_df_ss.shape)
data_res_a = pd.merge(stock_df_ss[['代码', '名称', '最新价', '今开']],
                      stock_zh_a_hist_df_last_day,
                      on=['代码'],
                      how="left")
data_res_a = pd.merge(data_res_a, data_avg_60, on=['代码'], how="left")
data_res_a.columns = ['代码', '名称', '最新价', '今开', '昨开', '昨收', '收盘_60']
```

```python
data_res_a["老A"] = (data_res_a["最新价"] - data_res_a["昨收"]) / (data_res_a["昨开"] - data_res_a["昨收"])
data_res_a["老A"] = data_res_a["老A"].round(2)
data_res_b = data_res_a[data_res_a['老A'].between(0.5, 1.5)]
print(data_res_b.shape)
```

```python
data_res_b["老C"] = data_res_b['最新价'] / data_res_b['收盘_60']
data_res_b["老C"] = data_res_b["老C"].round(2)
data_res_c = data_res_b[data_res_b['老C'].between(0.7, 1.2)]
print(data_res_c.shape)
data_res_c
```

```python
data_res_all = pd.merge(data_res_c,
                        data_list_df[["代码", "版块", "行业"]],
                        on=['代码'],
                        how="left")
data_res_all['日期'] = today
data_res_all = data_res_all[['代码', '名称', '版块', '行业', '日期', '最新价', '今开', '昨开', '昨收', '收盘_60', '老A', '老C']]
data_res_all = data_res_all.sort_values(['代码'])
data_res_all["代码"] = "\t" + data_res_all["代码"]
data_res_all.to_excel("data_res_all_" + today + ".xlsx", index=False)
data_res_all
```
