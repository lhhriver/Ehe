```python
import os
import re
```

```python
def sort_by_number(string_list):
    def extract_number(string):
        matches = re.findall(r'\d+', string)
        return int(matches[0]) if matches else 0

    sorted_list = sorted(string_list, key=lambda x: extract_number(x))
    return sorted_list
```

```python
file_dir = "../../"
file_dir = os.path.abspath(file_dir)

all_md_li = []
paths_li = []
dir_set = set()  # å½“å‰ç›®å½•

for root, dirs, files in os.walk(file_dir):
    root = root[15:].replace("\\", "/")

    if ".git" not in root and "N02-ICB" not in root:
        files = sort_by_number(files) # æŒ‰æ•°å­—æ’åº
        for file in files:
            if file not in ["readme.md", "README.md", "Readme.md"]:
                if os.path.splitext(file)[1] == '.' + "md":
                    dir_tmp_li = root.split("/")

                    # æ ‡é¢˜
                    for i in range(len(dir_tmp_li)):
                        if "&&".join(dir_tmp_li[:i+1]) not in dir_set:
                            dir_set.add("&&".join(dir_tmp_li[:i+1]))
                            all_md_li.append("#" * (i + 1) + " " + dir_tmp_li[i])
                            print("#" * (i + 1) + " " + dir_tmp_li[i])

                    # æ–‡ç« 
                    all_md_li.append("[ğŸ”– %s](%s)" % (os.path.splitext(file)[0], os.path.join(root, file).replace("\\", "/").replace(" ", "%20")))
                    paths_li.append(os.path.join(root, file).replace("\\", "/").replace(" ", "%20"))
            
```

```python
# æ€»ç›®å½•
with open(os.path.join(file_dir, "N13-DDU/N04-Docsify/IDEhe", "N08-all_in_one.md"), "w", encoding="utf-8") as f:
    for li in all_md_li:
        f.write(li+"\n\n")
```
