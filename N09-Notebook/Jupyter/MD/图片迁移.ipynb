# 图片迁移

```python
import os
import pandas as pd
import shutil

path = r"F:\Gitee"
obj_dir = r"F:\Gitee\Ehe\images"

if os.path.exists(obj_dir):
    pass
    # shutil.rmtree(obj_dir)
    # os.makedirs(obj_dir)
else:
    os.makedirs(obj_dir)

img_list = []


def img_rename(obj_dir, img_name):
    if os.path.exists(os.path.join(obj_dir, img_name)):
        pass
        img_name_new = os.path.splitext(img_name)[0] + '_1' + os.path.splitext(
            img_name)[-1]
        img_name = img_rename(obj_dir, img_name_new)
    return img_name


def filter_dir(di):
    st = [1 for i in filter_dirs if i in di]
    if st:
        return False
    else:
        return True


work_dirs = [r'F:\Gitee\March', r'F:\Gitee\Taishan']

filter_dirs = [
    r'F:\Gitee\March\Base\06-其它', r'F:\Gitee\Taishan\AC_Project',
    r'F:\Gitee\Taishan\March\深度学习课程', r'F:\Gitee\Taishan\Anchor\动手学深度学习',
    r'F:\Gitee\Taishan\WorkSpace'
]

img_list = []

for path in work_dirs:
    for root, dirs, files in os.walk(path):
        for di in dirs:
            if filter_dir(os.path.join(root, di)) and di == 'images':
                for img in os.listdir(os.path.join(root, di)):
                    img_new = img_rename(obj_dir, img)
                    shutil.copyfile(os.path.join(os.path.join(root, di), img),
                                    os.path.join(obj_dir, img_new))
                    img_list.append((os.path.join(os.path.join(root, di),
                                                  img), img_new))
```

# HTML迁移

```python
import os
import pandas as pd
import shutil


def filter_dir(di):
    st = [1 for i in filter_dirs if i in di]
    if st:
        return False
    else:
        return True


def modify_img_path(html_path):
    string_all = open(html_path, 'r', encoding='utf-8').readlines()
    for i in range(len(string_all) - 1):
        if 'images' in string_all[i]:
            string_all[i] = string_all[i].replace(
                './images', 'images').replace(
                    'images/',
                    'https://gitee.com/liuhuihe/Ehe/raw/master/images/')

    open(html_path, 'w', encoding='utf-8').writelines(string_all)


work_dirs = [r'F:\Gitee\March', r'F:\Gitee\Taishan']

filter_dirs = [
    r'F:\Gitee\March\Base\06-其它', r'F:\Gitee\Taishan\AC_Project',
    r'F:\Gitee\Taishan\HTML', r'F:\Gitee\Taishan\March\深度学习课程',
    r'F:\Gitee\Taishan\Anchor\动手学深度学习', r'F:\Gitee\Taishan\WorkSpace'
]

for path in work_dirs:
    for root, dirs, files in os.walk(path):
        for i in files:
            if os.path.splitext(i)[-1] == '.html' and filter_dir(
                    os.path.join(root, i)):

                root_new = root.replace(r"F:\Gitee", r"F:\Gitee\Taishan\HTML")
                print(root_new)

                if not os.path.exists(root_new):
                    os.makedirs(root_new)

                shutil.copyfile(os.path.join(root, i),
                                os.path.join(root_new, i))
                modify_img_path(os.path.join(root_new, i))
```
