# md批处理

````python
import os
import shutil
````

````python
file_path = "../md/"
file_li = ["../md/" + li for li in os.listdir(file_path) if li.endswith(".md")]
file_li
````

````python
for file in file_li:
    with open(file, 'r', encoding='utf-8') as f:
        data_li = f.readlines()
        
    data_li_new = []
    for line in data_li:
        if line[0] in ["图","#", "*" , '-'] or line[0].isdigit():
            data_li_new.append(line)
        elif "Files/Typora" in line:
            #print(line)
            data_li_new.append("\n")
        elif line.startswith(">"):
            data_li_new.append(line[1:].strip() + "\n")
        else:
            data_li_new.append(line.replace(" ", "").replace(" ","").replace(" ，","，"))
            
    with open(os.path.basename(file), "w", encoding='utf-8') as f:
        f.writelines(data_li_new)
````

````python
import os 
````

````python
md_file_path = r"C:\LocalSpace\Gitee\ICEhe\N03-SystemVerilog\N01-SystemVerilog验证【蓝三版】\SystemVerilog-代码示例"
md_file_dir_li = [os.path.join(md_file_path,li) for li in os.listdir(md_file_path)]
md_file_dir_li
````

````python
for file_path in md_file_dir_li:
    N = 0
    print(file_path)
    tmp = [li for li in os.listdir(file_path)]
    tmp_li = sorted([[li.split(".")[1], li] for li in tmp], key=lambda x: int(x[0]))
    file_path_li = [os.path.join(file_path, li[1]) for li in tmp_li]
    md_li_all = ["# " + os.path.basename(file_path) + "\n"]
    for file in file_path_li:
        N += 1
        md_li_tmp = ["\n\n", "## ", os.path.basename(file), "\n", "```systemverilog\n"]
        if str(N) != os.path.basename(file).split(".")[1]:
            print(file)
        with open(file, "r", encoding='utf-8') as f:
            data_file_tmp = f.readlines()

        md_li_tmp.extend(data_file_tmp)
        md_li_tmp.append("\n```")
        md_li_tmp.append("\n> ")
        md_li_all.extend(md_li_tmp)

    with open(os.path.basename(file_path) + ".md", "w", encoding="utf-8") as f:
        f.writelines(md_li_all)
````
