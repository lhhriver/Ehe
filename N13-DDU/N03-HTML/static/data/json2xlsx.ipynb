```python
import os
import json
import pandas as pd

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

# xlsx转js

```python
data_xlsx_df = pd.read_excel("all_data_code.xlsx", sheet_name=None)
data_info = data_xlsx_df["info"]
data_value = data_xlsx_df["value"]
```

```python
data_info
```

```python
data_value.head()
```

```python
all_cls_js = {}

# 处理info
for info in data_info.to_dict(orient = "record"):
    tmp_dt = {}
    for k, v in info.items():
        if k == "cls":
            all_cls_js[v] = {}
        elif k == "desc":
            tmp_dt[k] = v
        else:
            tmp_dt[k] = eval(v)
            
    all_cls_js[info["cls"]]["info"] = tmp_dt
    all_cls_js[info["cls"]]["value"] = {}
        
all_cls_js
```

```python
# 处理 value
for li in data_value.to_dict(orient = "record"):
    all_cls_js[li["cls"]]["value"][li["key"]] = li["value"].split("\n")
    
all_cls_js
```

```python
with open("all_data_code.json", "w", encoding="utf-8") as f:
    json.dump(all_cls_js, f, ensure_ascii=False)
```

```python
with open("all_data_code.js", 'w', encoding="utf-8") as f:
    f.write("all_data = ")
    json.dump(all_cls_js, f, ensure_ascii=False)
```

# json转xlsx

```python
with open("all_data_code.json", "r", encoding="utf-8") as f:
    data_js = json.load(f)
```

```python
data_js
```

```python
res_info = []
for cls, data in data_js.items():
    tmp = [cls]
    for k, v in data['info'].items():
        tmp.append(v)
    res_info.append(tmp)
    
res_info
```

```python
res_value = []
for cls, data in data_js.items():
    for k, v in data['value'].items():
        tmp = [cls]
        tmp.append(k)
        tmp.append("\n".join(v))
        res_value.append(tmp)
    
res_value
```

```python
data_info_df = pd.DataFrame(res_info, columns=["cls","desc","fontsize","top","left","animate_top"])
data_info_df.head()
```

```python
data_value_df = pd.DataFrame(res_value, columns=["cls", "key", "value"])
data_value_df.head()
```

```python
writer = pd.ExcelWriter("all_data_code.xlsx")
data_info_df.to_excel(writer, sheet_name="info", index=False, encoding="utf-8")
data_value_df.to_excel(writer, sheet_name="value", index=False, encoding="utf-8")
writer.save()
writer.close()
```
