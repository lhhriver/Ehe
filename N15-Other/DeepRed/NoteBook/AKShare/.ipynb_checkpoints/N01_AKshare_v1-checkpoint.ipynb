# 历史行情数据

## 历史行情数据-东财

```python
stock_zh_a_hist_df_60 = pd.DataFrame()
for code in stock_df_ss['代码'].values.tolist():
    stock_zh_a_hist_df = ak.stock_zh_a_hist(symbol=code, period="daily", start_date=b_60_day, end_date=last_day, adjust="")
    stock_zh_a_hist_df['代码'] = code
    stock_zh_a_hist_df_60 = stock_zh_a_hist_df_60.append(stock_zh_a_hist_df)
    
stock_zh_a_hist_df_60
```

```python
from threading import Thread, Lock, Semaphore
sem = Semaphore(10)  # 此变量用于控制最大并发数

stock_zh_a_hist_df_60 = pd.DataFrame()

def work(code,lock):
    global stock_zh_a_hist_df_60
    sem.acquire()
    stock_zh_a_hist_df = ak.stock_zh_a_hist(symbol=code, period="daily", start_date=b_60_day, end_date=last_day, adjust="")
    stock_zh_a_hist_df['代码'] = code

    lock.acquire()
    stock_zh_a_hist_df_60 = stock_zh_a_hist_df_60.append(stock_zh_a_hist_df)
    lock.release()
    sem.release()
        
stock_list = stock_df_ss['代码'].values.tolist()
lock = Lock()

l = [Thread(target=work, args=(code, lock)) for code in stock_list]

l = []
for eachlist in [stock_list[i:i + int(len(stock_list) / 100)] for i in range(0, len(stock_list), int(len(stock_list) / 100))]:
    p = Thread(target=work, args=(code, lock))
    l.append(p)

for p in l:
    p.start()
    
for p in l:
    p.join()
```
